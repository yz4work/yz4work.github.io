```python
import numpy as np 
# %matplotlib
import matplotlib.pyplot as plt 
from matplotlib.patches import Ellipse
```

```python
## 画椭圆、圆

def plot_ellipse(ax, xy, w, h, theta, alpha, color):
    """ xy: 圆心，w: x轴长度，h: y轴长度，theta: 旋转角度"""
    vmax = max(w, h) // 2
    ell = Ellipse(xy=xy, width=w, height=h, angle=theta)
    ax.add_artist(ell)

    # 设置透明度、颜色
    ell.set_alpha(alpha), ell.set_facecolor(color)

    ## 设置边框
    ax.set_xlim(-vmax-1, vmax+1), ax.set_ylim(-vmax-1, vmax+1)

    plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(8,4), subplot_kw={'aspect': 'equal'})
plot_ellipse(axes[0], (0,0), 4, 4, 45, 0.5, 'red')
```

#### 等价替换

```python
## x = z
f = x + exp(y)
f.subs(x, z) # 注释：subs是替换的缩写，substitute，把f里的x用z代替
```

```python
## 结合等价替换和函数展开
f = sin(2*x) + cos(2*y)
f.subs(sin(2*x), expand_trig(sin(2*x))) # 用展开式替换原来式子
```

```python
## x = z, y = ln(z)
f = x + exp(y)
f.subs([(x, z), (y, log(z))]) ## 注释：多个变量的替换
```

```python
## x = z, y = ln(z)
f = x + exp(y)
f.subs([(x, 0), (y, 0)]) ## 赋值
```

#### 微分、积分

```python
## 导数
f = sin(x) * y
diff(f, x) # 注释：diff是微分的意思，这里对x求导
```

```python
## 二次导数
f = sin(x) * y
diff(f, x, x) ## 对x求两次导
```

```python
f = sin(x) * y
diff(f, x, y) ## 先对x，再对y
```

```python
## 积分
f = sin(x) + x
integrate(f, x) # 注释：integrate是积分的意思，这里对x积分
```

```python
## 定积分
f = sin(x) + x
integrate(f, (x, -1, 1)) # 注释：对x求积分，在[-1, 1]之间
```

```python
f = sin(x**2)
integrate(f, (x, -oo, oo)) # 注释：对x求积分，在[负无穷, 正无穷]之间
```

```python
## 极限
f = sin(x) / x
limit(f, x, 0) # 注释：limit是极限的意思，x趋近0
```

```python
f = sin(x)
f.series(x, 0, 4)
```

#### 解方程

```python
## 解方程
f = x**2 + 2*x
solve(Eq(f, 0), x) # 注释：solve是求解的意思，f=0, x=?
```

```python
f = x**2 + 2*x
solve(Eq(f, -1), x) # 注释：设定f=-1
```

```python
f = x**2 + y**2 -2*x*y 
solve(Eq(f, 0), y)
```

```python
## 多个方程，经过(0, 0), (0, 2)，半径为1的圆心
eq1 = Eq((0 - x)**2 + (0 - y)**2, 1)
eq2 = Eq((0 - x)**2 + (2 - y)**2, 1)
vars = [x, y] ## 要求解的变量
nonlinsolve([eq1, eq2], vars)
```

```python
x1, x2, x3, x4, x5, x6 = symbols('x1, x2, x3, x4, x5, x6')
w0, w1 = symbols('w0, w1')

a1 = exp(w0+2*w1) / (1 + exp(w0+2*w1))
a2 = exp(-w0 - 7*w1) / (1 + exp(-w0 - 7*w1))
a3 = exp(-w0 - 4*w1) / (1 + exp(-w0 - 4*w1))
a4 = exp(w0 + w1) / (1 + exp(w0 + w1))
a5 = exp(-w0 - 5*w1) / (1 + exp(-w0 - 5*w1))
a6 = exp(w0 + 4.5*w1) / (1 + exp(w0 + 4.5*w1))

f1 = a1 - a2 - a3 + a4 - a5 + a6
f2 = 2*a1 - 7*a2 - 4*a3 + a4 - 5*a5 + 4.5*a6
```

```python
vars = [w0, w1] ## 要求解的变量
nonlinsolve([f1, f2], vars)
```

```python
f1 = x1/(1+x1)
f2 = x2/(1+x2)
solve([f1.subs(x1, a1), f2.subs(x2, a2)])
```
